```python
import pandas as pd
import numpy as npb
```

```python
data=pd.read_csv('/content/Mini Project - Marks Adding.csv')
```

```python
data.head()
```

```python
data=data.rename(columns={'Roll Num':'Roll_no'})
data.head()
```

```python
for Roll_no , marks in data.items():
  print(f"Roll Number : {Roll_no}, Marks are : {marks}")
```

```python
total_marks = {}
for Roll_no , marks in data.items():
  if Roll_no in total_marks:
    total_marks[Roll_no] +=marks
  else:
    total_marks[Roll_no]=marks
```

```python
for Roll_no , total in total_marks.items():
  print(f" ROll number : {Roll_no}, Total Marks: {total}")
```
